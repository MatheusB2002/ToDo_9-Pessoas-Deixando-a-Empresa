Uma boa conclusão após analisármos os dados foi que existem sim alguns problemas graves em relação à forma com que a empresa toma conta de seus funcionários. 
Foi possível observar que estamos perdendo uma quantidade considerável de funcionários muito capacitados na empresa, essa observação foi feita quando os gráficos demonstraram uma grande alta em saída por motivo de o funcionário estar avançando na carreira dentro da empresa mas não encontra oportunidades de usar esse conhecimento dentro, logo, ele sai para empresas que **reconhecem** a capacidade do que aquele conhecimento que ele tem pode entregar na empresa.

A palavra "reconhecem" não esta em **negrito** no parágrafo anterior atoa, esse é mais um problema grave que foi identificado e é um problema o quanto tanto triste já que o funcionário se dedica no trabalho e traz bons resultados à empresa e isso não é gratificado de forma nenhuma.

Isso deve acontecer pois a relação Funcionários x Empresa não está legal, o crucial seria a empresa fica mais próximo ao funcionário de alguma forma para que possa acompanhar a evolução, o desenvolvimento e as entregas dos funcionários para que fique mais agradável tanto para a empresa quanto para o bem estar do funionário.
